### 词云的绘制 地震高频地区分布词云图

```python
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# perpare and load data
name_data = pd.read_csv('Earthquake1.csv', usecols=['LOCATION_C'])
name_data
```

```python
# parse data
groupby_count = name_data.groupby('LOCATION_C').LOCATION_C.count()
sort_data = groupby_count.sort_values(ascending=False)
sort_data
```

```python
# plot
def analysis_address():
    plt.figure(figsize=(20, 8))
    font_path = 'SimHei.ttf'
    wordcloud = WordCloud( 
        background_color='#F3F3F3', # 背景颜色 
        width=1200, # 宽 
        height=800, # 高 
        font_path=font_path, # 字体
        max_font_size=200, # 最大的字体 
        random_state=42, # 颜色int or None 
        scale=2, # 清晰度 
        colormap='viridis', # 默认virdis 
    )
    
    wordcloud.fit_words(sort_data)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
analysis_address()
```
